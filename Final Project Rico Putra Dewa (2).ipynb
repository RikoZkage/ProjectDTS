### 3. Dari 2 features yang memiliki korelasi kuat dalam dataset ini yaitu hubungan antara Jumlah Full Liquid dan Total Konsumsi CO2 (Total), akan dibuat statistika deskriptif dari 2 features ini.

```python
# Statistika Deskriptif untuk features Liquid Fuel
import pandas as pd
import numpy as np

# menentukan dataframe dari Liquid Fuel
Lf = data['Liquid Fuel']

# mencari mean, median, modus, maksimal, minimal, jarak, variasi, simpangan baku dan koefisien variasi
meanLf = Lf.mean()
medianLf = Lf.median()
modusLf = Lf.mode()
maksLf = Lf.max()
minLf = Lf.min()
jarakLf = maksLf - minLf
variansLf = Lf.var()
simp_bakuLf = Lf.std()
koef_varLf = simp_bakuLf / meanLf

# menampilkan ukuran Liquid Fuel
print('Mean Jumlah Liquid Full : ',meanLf)
print('Median Jumlah Liquid Full : ',medianLf)
print('Modus Jumlah Liquid Full : ',modusLf[0])
print('Maksimum Jumlah Liquid Full : ',maksLf)
print('Minimum Jumlah Liquid Full : ',minLf)
print('Jarak Jumlah Liquid Full : ',jarakLf)
print('Varian Jumlah Liquid Full : ',variansLf)
print('Simpangan Baku Jumlah Liquid Full : ',simp_bakuLf)
print('Koefisien Variasi Jumlah Liquid Full : ', koef_varLf)
```

```python

import pandas as pd
import numpy as np

# menentukan dataframe dari Total Konsumsi CO2
To = data['Total']

# mencari mean, median, modus, maksimal, minimal, jarak, variasi, simpangan baku dan koefisien variasi
meanTo = To.mean()
medianTo = To.median()
modusTo = To.mode()
maksTo = To.max()
minTo = To.min()
jarakTo = maksTo - minTo
variansTo = To.var()
simp_bakuTo = To.std()
koef_varTo = simp_bakuTo / meanTo

# menampilkan ukuran Total Konsumsi CO2
print('Mean Total Konsumsi CO2 : ',meanTo)
print('Median Total Konsumsi CO2 : ',medianTo)
print('Modus Total Konsumsi CO2 : ',modusTo[0])
print('Maksimum Total Konsumsi CO2 : ',maksTo)
print('Minimum Total Konsumsi CO2 : ',minTo)
print('Jarak Total Konsumsi CO2 : ',jarakTo)
print('Varian Total Konsumsi CO2 : ',variansTo)
print('Simpangan Baku Total Konsumsi CO2 : ',simp_bakuTo)
print('Koefisien Variasi Total Konsumsi CO2 : ', koef_varTo)
```

### 4. Dari 2 Features yang memiliki korelasi kuat tersebut dimana bentuk dari pola sebaran yang terdapat pada scatter plot nya berupa regresi linier.

```python

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import statsmodels
import patsy
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
msk = np.random.rand(len(data)) < 0.8
train = data[msk]
test = data[~msk]
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['Liquid Fuel']])
train_y = np.asanyarray(train[['Total']])
regr.fit (train_x, train_y)
# The coefficients dan Intercept
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

membuat visual grafik perbandingan antara grafik dari hubungan 2 features yang terpilih untuk dibuat model regresi linier.

```python
plt.scatter(train.Year, train.Total,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Liquid Fuel")
plt.ylabel("Total")
```

akan ditunjukkan tingkat keakuratan berbagai metric yang digunakan untuk mengevaluasi model Regression dengan Scikit-learn. Dalam model ini akan digunakan diantara nya yaitu : Mean absolute error (MAN), Mean squared error (MSE), dan R2 Score.

Memilih 0.8 bagian data secara random untuk menghitung model serta menguji model tsb.

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['Liquid Fuel']])
test_y = np.asanyarray(test[['Total']])
test_y_ = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_ , test_y) )
```

- Membuat Scatterpot dari hasil regressi

```python
plt.scatter(test_x,test_y)
plt.plot(test_x, test_y, color='r')
plt.show()
```

Berdasarkan nilai dari R2 positif maka model ini dapat digunakan yang memiliki korelasi yang tepat. Demikian dikarenakan nilai dari R2 sebesar 0.97 dan itu merupakan nilai terbesar maka korelasi antara Luquid Full dan Total Konsumsi CO2 memiliki korelasi paling kuat.

Pra-pemrosesan
Semua features dalam dataset ini adalah variabel yang dikategorikal. Algoritma k-means tidak secara langsung berlaku untuk variabel kategori karena fungsi jarak Euclidean tidak benar-benar bermakna untuk variabel diskrit. Jadi, mari jatuhkan fitur ini dan jalankan pengelompokan.

Suatu Normalisasi atas deviasi standar dimana normalisasi menggunakan tandardScaler () untuk menormalkan dataset.

```python
from sklearn.preprocessing import StandardScaler
X = data.values[:,1:]
X = np.nan_to_num(X)
Clus_dataSet = StandardScaler().fit_transform(X)
Clus_dataSet
```

### Pemodelan K-Means pada Dataset

Untuk melihat apakah suatu objek berada di dalam cluster yang tepat, akan dimisalkan jumlah kluster tersebut dengan sembarang angka K, yang pertama misal nilai K=4.

```python
clusterNum = 4
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

Setelah mendapatkan label dari K-Means nya maka dimasukkan kedalam setiap baris pada dataframe

```python
data["Clus_km"] = labels
data.head
```

Setelah itu akan diperiksa nilai centroid dengan nilai rata-rata dari setiap features

```python
# menampilkan data tabel rata-rata untuk setiap Clus_km
data.groupby('Clus_km').mean()
```

Ada banyak cara untuk menentukan jumlah cluster salah satunya adalah dengan nilai Shilhouette Coefficient.  Shilhouette Coefficient berguna untuk mengetahui apakah suatu objek berada di dalam cluster yang tepat. Shilhouette Coefficient dihitung menggunakan jarak antar cluster dan jarak cluster terdekat untuk setiap sampel.

```python
# Menentukan suatu objeck berada dalam cluster yang tepat dengan Shilhouette Coefficient
from sklearn.metrics import silhouette_score

print(silhouette_score(X, k_means.labels_))
```

Untuk melihat perbandingan, akan digunakan jumlah kluster yang lainnya, dalam hal problem ini misal K=2.

```python
clusterNum = 2
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

```python
data["Clus_km"] = labels
data.head
```
